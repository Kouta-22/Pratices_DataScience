```python
vendas_mes.index
```

__Valor de media a ser testado__

```python
media = vendas_mes.mean()
```

```python
cores=[]
```

__Alimentando a lista de cores__

```python
for x in vendas_mes.values:
    if x < media:
        cores.append('red')
    else:
        cores.append('lightblue')
```

__Visualizando a lista de cores__

```python
cores
```

__Plotando grafico personalizado__

```python
data = [go.Bar(x=vendas_mes.index,
               y=vendas_mes.values,
               marker= {'color': cores,#Lista de cores
                        'line': {'color': '#333',
                                 'width': 2}
                                 },
                                 opacity=0.7
                )
        ]
# Criando Layout
config_layout = go.Layout(title='Vendas no Periodo',
                          xaxis={'title': 'Valores em Vendas'},
                          yaxis={'title': 'Periodo'})

# Objeto Figura
fig = go.Figure(data=data, layout=config_layout)

# Plotando Grafico
iplot(fig)
```

__Destacando a barra com maior valor e minimizando a visão de outras__

```python
maximo_vendas = vendas_mes.values.max()
```

```python
maximo_vendas
```

__Redefinindo as cores que serão usadas no grafico para gerar outros outliers__

```python



cores = []#limpando a lista de cores
for x,y in zip(vendas_mes.values,vendas_mes.index):
    if x == maximo_vendas:
        mes_maximo_vendas = y
        cores.append('purple')
    else:
        cores.append('lightgray')

```

```python
cores
```

```python
data = [go.Bar(x=vendas_mes.index,
               y=vendas_mes.values,
               marker={
                   'color': cores,
                   'line':{'color':'#333',
                            'width':2}
                       },
               opacity=0.7
               )
        ]
# Criando Layout
config_layout = go.Layout(title='Vendas no Periodo',
                          xaxis={'title': 'Valores em Vendas'},
                          yaxis={'title': 'Periodo'},
                          annotations = [{'text':'Mes Destaque de Vendas',#Texto na Barra de destaque
                                        'x':mes_maximo_vendas,
                                        'y':maximo_vendas}]
                          )

# Objeto Figura
fig = go.Figure(data=data, layout=config_layout)

# Plotando Grafico
iplot(fig)
```
