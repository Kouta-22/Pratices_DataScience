```python
import statistics as stats
from scipy.stats import kurtosis , skew


def calcula_estatistica(data):
    """
    Calculates various statistical measures for a given data set.

    Args:
        data (list): A list of numerical values.

    Returns:
        str: A formatted string containing the calculated statistics.
    """

    mean = round(np.mean(data), 2)
    mode = stats.mode(data)
    if isinstance(mode, list):
        mode = round(mode[0], 2)  # Acessa e arredonda o primeiro elemento se for uma lista
    else:
        mode = round(mode, 2)  # Arredonda o inteiro diretamente se não for uma lista
    median = round(np.median(data), 2)
    variance = round(np.var(data), 2)
    std_dev = round(np.std(data), 2)
    skewness = round(skew(data), 2)  # Calculate skewness
    kurtosis_value = round(kurtosis(data), 2)

    message = (
        f"Mean: {mean}\n"
        f"Mode: {mode}\n"
        f"Median: {median}\n"
        f"Variance: {variance}\n"
        f"Standard Deviation: {std_dev}\n"
        f"Skewness: {skewness}\n"
        f"Kurtosis: {kurtosis_value}"
    )

    return message


# Example usage
data = [4, 5, 6, 5, 5, 6, 6, 5, 4, 5, 5, 5]
result = calcula_estatistica(data)
print(result)
```

```python
ms0 = [4,5,6,5,5,6,6,5,4,5,5,5]
```

```python
calcula_estatistica(ms0)
```

Distribuição Simetrica - Media = Moda = Mediana

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig= plt.figure(figsize=(12, 4))  # Create figure and axis

# Create the distribution plot
ax = sns.distplot(ms0, rug=True, kde=True, hist=False)

# Add annotations after the plot is generated
ax.annotate('Mode', xy=(5, 0.4), xytext=(4.7, 0.7), arrowprops={'arrowstyle': '-'})
ax.annotate('Median', xy=(5, 0.4), xytext=(5.3, 0.7), arrowprops={'arrowstyle': '-'})
ax.annotate('Mean' , xy=(5, 0.4), xytext=(5, 0.1), arrowprops={'arrowstyle': '-'})

plt.show()
```
