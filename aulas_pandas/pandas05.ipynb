# Tabelas Pivot
- tabelas pivot são uteis para fazer agrupamento nos dados
- o parametro index informa as colunas que serao usadas para agregação
- o parametro aggfunc é usado para definir a função de agregação que no caso é count(), mas poderia ser media por exemplo
- o parametro margins = True calcula a quantidade total no final da tabela


```python
#Conta a quantidade de imoveis agrupados pelas colunas waterfront e floors
import pandas as pd
arquivo = 'kc_house_data.csv'
dataset = pd.read_csv(arquivo,sep=',',header=0)
```

```python
dataset.pivot_table('id',index=["waterfront","floors"], aggfunc='count', margins=True)
# O codigo acima faz uma tabela que mostra os a quantidade de andares(floors) de uma casa com vista para o mar(waterfront)
```

# Croostab
- Essa funcionalidade cruza valores das variaveis
- Nesse exemplo podemos ver qual a distribuição dos imoveis por numero de quartos com relação a sua condição

```python
pd.crosstab(dataset['bedrooms'],dataset['condition'])
#O codigo acima faz uma tabulação cruzada entre as condições (condition) para as casas com determinadas quantidade de quartos
# 1 para condição ruim 5 para condição otima
```

```python
table = pd.crosstab(dataset['bedrooms'], dataset['condition'])
table.plot(kind='bar', width=1.0, color=['red','yellow','orange','blue','green'], title='Condition By Bedrooms',grid=False)
```
