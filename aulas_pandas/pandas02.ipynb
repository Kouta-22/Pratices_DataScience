# Consultando um Dataset
- podemos fazer consultas em um dataframe que é semelhante a linhagem SQL
- Existem métodos interessantes para fazer consultas usando operadoeres lógicos (>,<,==)
- Pode-se fazer consultas usando instruções de agrupamento

```python
import pandas as pd
```

```python
arquivo = 'kc_house_data.csv'
```

```python
dataset = pd.read_csv(arquivo, sep=',', header = 0)
```

```python
pd.value_counts(dataset['bedrooms'])
#conta a quantidade de valores unicos
```

```python
#O metodo loc() é usado para visualizar infomações do dataset.
# Este metodo recebe uma lista por paramento e retorna o resultado da consulta
# Consulta imoveis com 3 quartos
dataset.loc[dataset['bedrooms']==3]
```

```python
#Usando o metodo loc junto com operador &
# Consulta imóveis com 3 quartos e com numero de banheiros maior que 2
dataset.loc[(dataset['bedrooms']==3) & (dataset['bathrooms'] > 2)]
```

```python
# O metodo sort_values() ordena o dataset pela coluna 'price' em ordem descresente
# ATENÇÂO apenas o retorno da query será ordenado, não fara organização do dataset
dataset.sort_values(by='price', ascending=False)
#dataset = dataset.sort_values(by='price', ascending=False) || para fazer a alteração do dataset em memoria
```

```python
dataset[dataset['bedrooms']==4].count()
```

# Alterando dataset

```python
dataset.head()
```

```python
# Adicionar uma coluna ao Dataframe
dataset['size'] = (dataset['bedrooms']*20)
```

```python
dataset['size'].head()
```

```python
#Criando função
def categoriza(s):
    if s >= 80: return 'Big'
    elif s >= 60: return 'Medium'
    elif s >= 40: return 'Small'
#Função criada

```

```python
# Criando uma nova coluna a partir do processamento realizado e chamando a função criada anteriormente
dataset['cat_size'] = dataset['size'].apply(categoriza)
```

```python
dataset['cat_size']
```

```python
pd.value_counts(dataset['cat_size'])
#ver a distribuição da coluna com o metodo value_counts.
```

```python
# O metodo drop é usado para excluir dados de um dataframe
# A opção axis=1 define que queremos excluir uma coluna e não uma linha
#O paramentro inplace define que a alterção ira modificar o objeto em memoria
dataset.drop(['cat_size'], axis= 1, inplace=True)
```

```python
dataset.head()
```

```python
#Caso não utilize o parametro inplace o pandas considera como False e não apagara na memoria apenas na query
dataset.drop(['size'], axis=1, inplace=True)
```

Apagando linhas baseado em condições logicas

```python
# Dropa(apaga) linhas com bedrooms = 0
dataset.drop(dataset[dataset.bedrooms == 0 ].index,inplace=True)
```

```python
dataset.drop(dataset[dataset.bedrooms>30].index, inplace=True)
#pd.value_counts(dataset['bedrooms']) foi apagado os quartos cm tamanho maior que 30
```

# Percorrendo as linhas de uma dataframe
- O metodo iterrows() permite percorrer por todas as linhas de uma dataframe
- Esse metodo retorna um objeto **iterator** que contem o indice de cada linha e um cada linha dado tipo série

```python
type(dataset.iterrows())
```

```python
next(dataset.iterrows())
```

```python
# percorrendo o dataframe e imprimindo o indice de cada linha
for indice, linha in dataset.head(1).iterrows():
    print(indice,linha)
```

```python
for indice, linha in dataset.head(5).iterrows():
    print(indice,linha['bedrooms'], linha['floors'], linha['price'])
```

```python
dataset.price.head()
```

```python
#percorrendo e atulizando linhas de um dataframe
#atualizando o valor da coluna PRICE multiplicando seu valor por 2
#è preciso utilizar o metodo at()
for indice, linha in dataset.iterrows():
    dataset.at[indice, 'price'] = linha['price'] *2
```

```python
dataset.price.head()
```

# Percorrendo um dataframe com o metodo itertuples()
- retorna as linhas e indice em formato de tuplas
- costuma ser mais rapido que o iterrows()

```python
#percorre o dataframe usando itertuples()
for linha in dataset.head().itertuples():
    print(linha)
```
