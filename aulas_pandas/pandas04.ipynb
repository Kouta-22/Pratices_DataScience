# BoxPlot
- Graficos de boxplot pode ser horizontal ou vertical com duas retas na parte supeior de inferior
- O retângulo é formado por três **Quartis** que dividem os dados em rols com 25% dos dados cada
- O Quartil 1 é representado o valor medio dos 25% dos dados
- O Quartil 2 é a mediana dos valores pela linha de borda superior do retângulo que representa o valor medio dos 75% dos dados
- As linhas **superior** e **inferiror** extra ao retangulo representam o valor máximo e minimo dos dados

# extraindo Insights
- Os graficos de boxplot são excelentes ferramentas de analise de dados, principalmente para identificar **outliers**
Um outlier é um valor que foge da normalidade e que pode (e provavelmente irá) causar anomalias nos resultado de análise

```python
import pandas as pd
import matplotlib
import matplotlib.style
arquivo = 'kc_house_data.csv'
dataset = pd.read_csv(arquivo, sep=',',header=0)
```

```python
# Plotando grafico do tipo boxplot da coluna bedrooms
# Esse grafico possui muito outliers
%matplotlib inline

matplotlib.style.use('classic')
dataset.boxplot(column='bedrooms')
```

```python
# Visualizando os estilos de layout disponivel
matplotlib.style.available
```

```python
dataset.boxplot(column='price', by='bedrooms')
```


- A função errors='coerce' é um parâmetro opcional da função pd.to_numeric() do Pandas. Quando você utiliza esse parâmetro, ele especifica como lidar com valores que não podem ser convertidos para numéricos.

- Quando errors='coerce' é definido, qualquer valor que não possa ser convertido para numérico será tratado como NaN (Not a Number), ou seja, como um valor ausente.

```python
# Converter todas as colunas para tipos numéricos, tratando valores não numéricos como NaN
dataset = dataset.apply(pd.to_numeric, errors='coerce')

# Calcular a correlação entre as colunas numéricas
correlation = dataset.corr()

# Exibir a correlação
print(correlation)

```
