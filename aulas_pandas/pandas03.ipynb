# Missing values
- **Missing** values são valores faltantes em colunas, esses podem ser oriundos de falhas em cargas de dados, dalhas em crawlers ou até mesmo corrupção de dados
- Missing values podem ser um problema em varias situações como por exemplo algoritmos de machine learning que não tabalham bem com dados faltantes 
- Estes também podem atapalhar resultados de analises

```python
import pandas as pd
```

```python
arquivo = 'kc_house_data.csv'
dataset = pd.read_csv(arquivo, sep=',', header = 0)
```

```python
#consulta as linhas com valores faltantes
dataset.isnull().sum()
```

```python
#Com esse comando removemos todas as linhas onde tenha pelo menos um registo faltante em algum atributo
dataset.dropna(inplace=True)
```

```python
dataset.dropna(how='all', inplace=True)
#è possivel ainda remover somente linhas que estejam com valores faltantes em todas as colunas
```

```python
#preenche com a media dos valores da coluna floors os valores null
#função "mean()" calcula a media
dataset['floors'].fillna(dataset['floors'].mean(), inplace=True)
```
