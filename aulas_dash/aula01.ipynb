```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import datetime
```

```python
df = pd.read_csv('C:/Users/Guilherme/OneDrive/√Årea de Trabalho/Pratices_DataScience/aulas_dash/data/sales_data_sample.csv', encoding='latin1')
```

```python
df.head()
```

```python
#Transformar em datetime
df['Data_Pedido'] = pd.to_datetime(df['ORDERDATE'])
df['Ano'] = df['Data_Pedido'].dt.year
df['Mes'] = df['Data_Pedido'].dt.month_name()

#Deletar colunas que n√£o serao usadas
drop_cols = ['ADDRESSLINE1', 'ADDRESSLINE2', 'POSTALCODE','CITY', 'TERRITORY', 'PHONE','STATE', 'CONTACTFIRSTNAME','CONTACTLASTNAME','CUSTOMERNAME','ORDERNUMBER']
df = df.drop(drop_cols,axis=1)

```

```python
#Agrupar dados por pais
sales_by_country_series = df.groupby('COUNTRY')['SALES'].sum()


```

__Iniciando a aplica√ß√£o e dando nome a ela__

```python
app = dash.Dash()#Iniciando aplica√ß√£o
app.title = 'Dashboard'
```

```python
app.layout = html.Div(
    html.Div([
        html.H1(children='Primeira Vizualiza√ß√£o'),

        html.Div(children='''
            Dashboard de Vendas
        '''),

        dcc.Graph(
            id='vendas-por-pais',
            figure={
                'data': [
                    {'x': sales_by_country_series.index, 'y':sales_by_country_series.values , 'type': 'bar', 'name': 'VbyC'}
                ],
                'layout': {
                    'title': 'Vendas Por Pa√≠s [$]'
                }
            }
        ),

    ])
)
```

```python
#Agrupando dados por ano
sales_by_year_series = df.groupby('Ano')['SALES'].sum()
```

```python
app.layout = html.Div(
    html.Div([
        html.H1(children='Primeira Vizualiza√ß√£o'),

        html.Div(children='''
            Dashboard de Vendas 
        '''),

        dcc.Graph(
            id='vendas-por-pais',
            figure={
                'data': [
                    {'x': sales_by_country_series.index, 'y':sales_by_country_series.values , 'type': 'bar', 'name': 'V-Por-P'}
                ],
                'layout': {
                    'title': 'Vendas Por Pa√≠s [$]',
                    'xaxis': dict(
                        title = 'Paises',
                        titlefont = dict(
                        family = 'Courier New, monospace',
                        size = 18,
                        colors='#7f7f7f'
                        )),
                    'yaxis' : dict(
                        title = 'Vendas [$]',
                        titlefont=dict(
                            family='Helvetica, monospace',
                            size=18,
                            colors='#7f7f7f'
                        ))
                }
            }
        ),

    dcc.Graph(
            id='grafico-2',
            figure={
                'data': [
                    {'x': sales_by_year_series.index, 'y':sales_by_year_series.values , 'type': 'bar', 'name': 'V-Por-P'},
                    {'x': sales_by_year_series.index, 'y':sales_by_year_series.values , 'type': 'line', 'name': 'V-Por-P'}
                ],
                'layout': {
                    'title': 'Vendas Por Pa√≠s [$]',
                    'xaxis': dict(
                        title = 'Paises',
                        titlefont = dict(
                        family = 'Courier New, monospace',
                        size = 18,
                        colors='#7f7f7f'
                        )),
                    'yaxis' : dict(
                        title = 'Vendas [$]',
                        titlefont=dict(
                            family='Helvetica, monospace',
                            size=18,
                            colors='#7f7f7f'
                        ))
                }
            }
        ),
    ])
)
```

__Iniciar o servidor da aplica√ß√£o Dash__
- __Parametros:__
    - App.run_server() √© um metodo fornecido pelo frameworck Dash para executar o servidor web para aplica√ß√£o Dash
    - debug = True habilita o modo de depura√ß√£o automatica
    - port = 8080, especifica a porta na qual o servidor ira fazer a requisi√ß√£o
    - use_reloader = False, desabilita o carregamento automatico

```python
if __name__ == '__main__':
    app.run_server(debug=True, port=8080, use_reloader=False)
```
