# Data View BoxPlot
- Graficos Boxplot é uma ferramenta grafica usada para representar a distribuição de dados numericos atraves de __QUARTIS__, eles são uteis para visualizar a dispersão e a simetria dos dados, bem como para identificar valores discrepantes(__OUTLIERS__)

__Um boxplot é composto pelos seguintes elementos__
- Minimo: O valor minimo do conjunto de dados, excluindo __outliers__
- Primeiro Quartil(Q1): O valor abaixo do qual se encontram 25% dos dados
- Segundo Quartil(Q2): O valor central do conjunto de dados. 50% dos dados são maiores que esse valor e os outros 50% dos dados são menores
- Terceiro Quartil(Q3): O valor abaixo do qual se encontram 75% dos dados
- Maximo: O valor maximo do conjunto de dados, excluindo __outliers__
- Outliers: Valores que estão fora de um intervalo definido, sendo considerados muito altos ou muito baixos

```python
#BOXPLOT de dias da semana por total de conta
import seaborn as sns
df_tips = sns.load_dataset("tips")
```

```python
sns.catplot(x="day",y="total_bill", kind='box',data=df_tips)
```

__Grafico de Boxplot dos dias por total de conta e se a pessoa é fumante ou não__

```python
sns.catplot(x="day",y="total_bill", kind="box",hue="smoker",data=df_tips)
```

```python
sns.catplot(x="day",y="total_bill", kind="boxen",data=df_tips)#a função BOXEN é mais focada para ver a distribuição massiva dos dados não da muita visão sobre outliers
```
