# Grafico de Linhas
- Grafico normalmente utilizado para visualização de dados em uma linha do tempo
- Normalmente utilizado com o eixo X relacionado a dados temporais e eixo Y a dados quantitativos

```python
# Eixo_x, Eixo_y
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
# Eixo_x, Eixo_y
plt.plot([1,2,3,4],[1,4,9,100])
plt.show()
```

__Definindo label aos eixos__

```python
# Eixo_x, Eixo_y
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
```

__Alterando a forma dos pontos__

```python
# Eixo_x, Eixo_y
plt.plot([1,2,3,4],[1,4,9,16],'o')
plt.show()
```

__Definindo limites de visualização dos eixos x e y__

```python
# Eixo_x, Eixo_y
plt.plot([1,2,3,4],[1,4,9,16],marker = 'o')#Por padrão é impresso a linha com o comando marker 
plt.axis([0,6,0,20])          #[xmin, xmax, ymin, ymax]
plt.show()
```

__Plotando dados categoricos__
- Praticando graficos em barra
- Ideal para quando estamos trabalhando com __comparação de valores__
- Normalmente quer se visualizar frequencia ou percentual de valores

```python
#Plotando graficos de barras verticais o Grafico de colunas
grupos = ['Produto A','Produto B', 'Produto C']
          #A  B   C
valores = [50,10,100]
        #Eixo X Eixo Y
plt.bar(grupos,valores)
plt.show()
```

# Plotando graficos de Barras juntas

```python
#Quantidade de vendas para o Produto A
valores_produto_A = [6,7,8,4,4]

#Quantidade de vendas para o Produto B
valores_produto_B = [3,12,3,4.1,6]

#Criando eixo X para produto A e produto B com uma separação de o.25 entre as barras
x1 = np.arange(len(valores_produto_A))
x2 = [x + 0.25 for x in x1]

#Plotando as barras
plt.bar(x1,valores_produto_A, width=0.25, label = 'Produto A', color = 'b')
plt.bar(x2,valores_produto_B,width=0.25, label = 'Produto B',color = 'g')

meses = ['Agosto','Setembro','Outubro','Novembro','Dezembro']
plt.xticks([x + 0.25 for x in range (len(valores_produto_A))],meses)

plt.legend()
plt.title("Quantidade de vendas")
plt.show()

```

__Customizando__
- Adicionado informações de percentual, sombra, separação das partes e legenda

```python
vendas = [3000, 2300,1000,500]
labels = ['E-commerce','Loja Fisica', 'E-mail','Marketplace']

#Define o nivel de separabilidade entre as partes, ordem do vetor representa as partes
explode = (0.1,0,0,0)
plt.pie(vendas, labels= labels, autopct='%1.1f%%', shadow= True, explode=explode)
plt.legend(labels,loc = 2)#Onde a legenda ira aparecer

#Define que o grafico sera plotado em circulo
plt.axis('equal')
plt.show()
```

# Matplotlib, pylot e pylab
- Pylot é um modulo do pacote matplotlib que implicitamente cria figuras e eixos nos graficos
- Pylab é outro modulo do matplotlib que já importa o numpy e pyplot. Este modulo é projetado para plots de arrays e calculos matematicos
- Pylab esta depreciado

```python
# NP foi importado mais cedo
import matplotlib.pyplot as plt

#Criando figura explicitamente
fig = plt.figure()

# Definindo Eixos
ax = fig.add_subplot(111)

# Scatter plot usando numpy
ax.scatter(np.linspace(0,1,5), np.linspace(0,5,5))

# Show the plot
plt.show()
```

__Acessando os subplots -> usa-se o metodo no singular "subplot"__

```python
x = [1,2,3,4]
y = [10,20,25,40]
plt.subplot(1,2,1)
plt.plot(x,y,'o')
plt.subplot(1,2,2)
plt.plot(x,y,color='r')
```

__Criando a intancia da classe figure__
- Aumenta o controle de nivel de costumização das funções do Matplotlib
- Gera menos codigo devido a orientação a objeto

```python
fig = plt.figure()#Criando um objeto do tipo figure

#Define o tamno em percentual do grafico
#Altura e largura do eixo na figura
axes_a = fig.add_axes([0.1,0.1,0.8,0.8])

#Plota o grafico
axes_a.plot(x,y)
```

__Criando uma figura com dois graficos de proporções diferentes__

```python
#Cria um objeto do tipo figure
fig = plt.figure()

#Define o tamno em percentual do grafico
#Altura e largura do eixo na figura
axes_a = fig.add_axes([0.1,0.1,0.8,0.8])
axes_b = fig.add_axes([0.2,0.5,0.3,0.3])

axes_a.plot(x,y)
axes_b.plot(x,y,'r--')#Linha tracejada em vermelho

```

__Customizando a largura e altura do grafico para acomodar melhor os dados__

```python
#x = [1,2,3,4]
#y = [10,20,25,40]

# Altera a largura e altura do grafico (width,heigth)
fig, ax = plt.subplots(figsize = (20,5))
ax.plot(x,y,'r--')#plota o grafico
```

__Plotando dois graficos(eixos) na mesma figura__

```python
#Define as configurações dos plots
#Cada plot tera o mesmo tamanho de figuras

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))

#Dados para cada subplot
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])

plt.show()
```

# Adicionando textos em graficos

```python
#Karnel resetado
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#Valores num intervalo de 0 a 2
x = np.linspace(0,2, num = 100)

#Define a legenda e o tamanho das linhas
plt.plot(x,x, label = 'linear')
plt.plot(x,x ** 2 , label = 'quadratico', linewidth = 5)

plt.xlabel('X label')
plt.ylabel('Y label')

plt.title('Grafico de linha simples')

#Exibe a legenda e por padrão usa o label de cada plot.
plt.legend()

#Configuração do texto.
plt.text(1.00,1.0,"Cruzamento de linhas", fontsize=10, horizontalalignment = 'right')

plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores num intervalo de 0 a 2
x = np.linspace(0, 2, num=100)

# Define a legenda e o tamanho das linhas
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratico', linewidth=5)

# Pontos de cruzamento
cross_points = [(0, 0), (1, 1)]
for point in cross_points:
    plt.plot(*point, 'ro')  # Desenha um ponto vermelho
    plt.text(point[0], point[1], f'  Cruzamento em {point}', verticalalignment='bottom')

plt.xlabel('X label')
plt.ylabel('Y label')

plt.title('Grafico de linha simples')

# Exibe a legenda e por padrão usa o label de cada plot.
plt.legend()

plt.show()
```
