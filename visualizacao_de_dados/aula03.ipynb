```python
import seaborn as sns
```

```python
df_tips = sns.load_dataset("tips")# É um dataset que ja vem por padrão na biblioteca seaborn
```

# Vizualizando a Relação Estatistica nos Dados
 - A analise estatistica é o processo de entender como variaveis no conjunto de dados se relacionam uma com a outra e como essas realções depende de outras variaveis. A visualização de dados se torna uma peça essencial nesse processo, pois, quando os dados são visualizados adequandamente, as pessoas conseguem descobrir padrões que indicam relações nos dados

# Função Replot()
- Metodo de alto nivel para visualização de relações estatisticas entre variaveis. O metodo replot( já combina o __Scatter Plot__ e __Line Plots__)
- Muito util para ver relações entre duas ou mais variaveis com uso de paramentros de semantica dos dados

# Relação de Variaveis com scatter plots
__Visualizando a relação entre a variavel total da conta e a gorjeta__

```python
sns.relplot(x='total_bill', y='tip',data=df_tips)# O dataframe ja foi carregado para variavel acima
```

__Visualizando a relção entre as variaveis total da conta e gorjeta e colorindo cada ponto com a variável smoker(fumante)__

```python
sns.relplot(x = "total_bill", y="tip",hue="smoker", data= df_tips);# É feita uma correlação entre as 3 variaveis 
                                                                  # Onde mostra o total da conta indicando quanto maior a conta maior a gorjeta e explicitando quem são os fumantes
```

__Correlação entre total da conta/gorjeta paga e o sexo__

```python
sns.relplot(x = "total_bill", y="tip",hue="sex", data= df_tips);
```

Stylizando a plotagem

```python
sns.relplot(x="total_bill",y="tip", hue="sex", style="sex", data=df_tips)
```

__Representando 4 variaveis e alterando a tonalidade e o estilo de cada ponto para visualizar insigths__

```python
sns.relplot(x="total_bill",y="tip", hue="sex", style="smoker", data=df_tips)#ATENÇÂO ao utilizar muitas variaveis para nao ficar confuso
```

# Cores no seaborn 
- A função que controla as cores no seaborn é a color_palette()
- Essa função fornece uma interface para gerar cores no seaborn
- Usa-se esse metodo internamente ao usar o parametro palette
- O metodo color_palette() recebe o nome de uma paleta de cores do seaborn ou um colormap do matplotlib
- O metodo set_palette() recebe os mesmos parametros e eh usado para definir um estilo padrão para todos os plots
- As cores devem ser definidas levando em consideração o tipo de dados que queremos plotar (qualitativo, sequenciais, divergentes)

__Paleta de cores para dados divergentes__

```python
df_tips = sns.load_dataset("tips")# Importando dataset
```

```python

sns.scatterplot(x="total_bill", y="tip", hue = "size", palette="Blues", size="size",sizes=(12,250),data=df_tips)
```

__Grafico utilizando a paleta de cor qualitativa "Set2"__
- Paleta de cor ideal para plotar dados categoricos

```python
sns.catplot(x="sex", kind="count", palette="Set2", data= df_tips)
```

__Definindo cores padrão para graficos Searborn__

```python
sns.set_palette("pastel")
#Plot usando a paleta de cor "paster"
sns.regplot(x="total_bill",y="tip",data=df_tips)
```

__Intrução With para definir cor temporariamente__

```python
with sns.color_palette("Accent"):
    sns.scatterplot(x="total_bill",y="tip",size="size",sizes=(12,250),data=df_tips)
```

# Utilizando o metodo catplot()
- O catplot() fornece acesso a várias funções de nivel de eixos que mostram a relação entre uma variavel numerica e uma ou mais variaveis categoricas usando uma de varias representação visuais
- Tipos de graficos catplot() permite plotar os seguintes tipos de visualizações modificados pelo parametro __KIND__:"strip","swarm","boxen","point","bar" ou "count"

```python
#Grafico de Colunas
sns.catplot(x="sex", kind="count", palette="Set2",data=df_tips)
#Eixo X = horizontal
#Eixo Y = vertical
```

```python
sns.catplot(y="day", kind="count", palette="Set1",data=df_tips)
```

__Scatter plot com dados categoricos__

```python
sns.catplot(x="day",y="total_bill", palette="Set2",data=df_tips)
```

O parametro swarn evita sobreposição dos pontos

```python
sns.catplot(x="day",y="total_bill",palette="Set1", kind='swarm',data=df_tips)
```

__O Parametro HUE permite adicionar uma terceira variavel para a visualização__

```python
sns.catplot(x="day",y="total_bill",hue="sex", kind='swarm',palette="Dark2",data=df_tips)
```
