# Histogramas e bins
- Um histograma é um tipo de grafico que representa a distribuição de frequencias de valores continuos ou discretos. Ele divide o eixo x em intervalos, conhecidos como bins e conta a quantidade de valores que caem em cada intervalo. A altura de cada barra representa a frequencia(quantidade) de valores dentro de cada intervalo(bins)

```python
import seaborn as sns
import numpy as np
df_tips = sns.load_dataset("tips")
```

```python
#Por padrão o metodo de histograma tenta encontrar a melhor quantidade de bins
#Mas pode ser definido manualmente
sns.set(color_codes=True)
```

```python
sns.displot(df_tips.total_bill,kde=False)#Usar o histplot o displot esta em desuso
```

```python
x = np.random.normal(size = 100) # x = np.random.randint(low=1, high=100, size=100) Para gerar numeros inteiros
                                    # x = np.random.randint(1,100, size =100)
```

Plotando o Histograma com bins = 20 e KDE = TRUE

```python
sns.histplot(df_tips.total_bill, bins = 15, kde=True)#O parametro KDE é esse linha que significa Kernel Density Estimation
```

# Jointplot (Scatterplot e Histograma)
- União de graficos do tipo scatter plot e histograma para duas variaveis 


```python
sns.jointplot(x="tip",y="total_bill",color='r',data=df_tips)
```

__Hexbin Plots__
- Adequado para visualização de grandes conjuntos de dados
- Tende a funcionar melhor com cores em branco de fundo

```python
sns.jointplot(x="tip",y="total_bill", color='b', kind="hex", data=df_tips)#Esse kind do tipo hex mostra uma maior densidade dos dados nos pontos mais escuros
```

__Usando o KDE para plotar mais de uma variavel__

```python
sns.jointplot(x="tip",y="total_bill",palette="Blues",kind="kde",data=df_tips)
```

# Vizualizando relacionamento de dados emparelhados
